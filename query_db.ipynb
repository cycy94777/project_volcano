```python
import sqlalchemy as sql
import json
```

```python
engine=sql.create_engine('sqlite:///data/db.sqlite')
results=engine.execute('select * from data').all()
results
```

```python
# Convert list of tuples to list of dictionaries
keys = ['id', 'name', 'country', 'type', 'last_eruption', 'age', 'region', 'continent', 'latitude', 'longitude', 'elevation', 'composition', 'tectonic_settings']
data_dict = [dict(zip(keys, values)) for values in results]

# Convert list of dictionaries to JSON format
json_data = json.dumps(data_dict)

# Print JSON data
print(json_data)
```
